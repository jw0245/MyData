```python
import pandas as pd
### 타이타닉 데이터불러오기
data = pd.read_csv('dataset/titanic_train.csv')
data.head(10)
```

```python
# 데이터의 존재여부를 위해 결측치 확인
data.isnull().sum()
```

```python
from faker import Faker
fake = Faker('ko-KR')
# 성병에 따른 이름 분류
def Sex(row):
    if row['Sex'] == 'female':
        new_name = fake.name_female()
    else:
        new_name = fake.name_male()
    return new_name

data['Name'] = data.apply(Sex, axis=1)

data.head(10)
```

```python
data.to_csv('change_train.csv', index=False)
```

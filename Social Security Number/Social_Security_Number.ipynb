```python
import pandas as pd

# 데이터 개수
n = 10
# 가명처리 문자열
change = 'XXXXXX'
ssn  = [["960111-1000000"] for _ in range(n)]

# 데이터 str 타입으로 변환
ssn = [list(map(str,ssn[i])) for i in range(len(ssn))]

df = pd.DataFrame(ssn, columns=['주민번호'], index=[i for i in range(0, len(ssn))])
df
```

```python
for i in range(len(df['주민번호'])):
    
    df['주민번호'][i] = df['주민번호'][i][0:8] + change
df
```
